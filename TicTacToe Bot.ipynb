```python
def initialDisplay(board, boardLength):
    count = 1
    for i in range(boardLength):
        row = []
        for j in range(boardLength):
            row.append(count)
            count = count + 1
        print(row)
```

```python
def createBoard(boardLength):
    board = dict()
    for i in range(int(boardLength**2)):
        board[i+1] = '#'
    return board
```

```python
def displayBoard(board, boardLength):
    count = 1
    for i in range(boardLength):
        row = []
        for j in range(boardLength):
            row.append(board[count])
            count = count + 1
        print(row)
```

```python
def moveCheck(board,boardLength,playerMove):
    if playerMove < 1 or playerMove > boardLength**2:
        print("That move is out of bounds.")
        return False
    if board[playerMove] != '#':
        print("That spot is already filled.")
        return False
    return True
```

```python
def findRow(board,boardLength,playerMove):
    rowChar = []
    rowNum = []
    layer = int((playerMove-1)/boardLength)
    for i in range(boardLength):
        rowChar.append(board[(boardLength*layer)+1+i])
        rowNum.append((boardLength*layer)+1+i)
    return rowChar, rowNum
```

```python
def findCol(board,boardLength,playerMove):
    colChar = []
    colNum = []
    tempMove = playerMove%boardLength
    if tempMove == 0:
        tempMove = boardLength
    for i in range(boardLength):
        colChar.append(board[tempMove+(i*boardLength)])
        colNum.append(tempMove+(i*boardLength))
    return colChar, colNum
```

```python
def findDiagonalPos(board,boardLength):
    diaPosChar = []
    diaPosNum = []
    for i in range(boardLength):
        diaPosChar.append(board[boardLength + (boardLength-1)*i])
        diaPosNum.append(boardLength + (boardLength-1)*i)
    return diaPosChar, diaPosNum
```

```python
def findDiagonalNeg(board,boardLength):
    diaNegChar = []
    diaNegNum = []
    for i in range(boardLength):
        diaNegChar.append(board[1 + (boardLength+1)*i])
        diaNegNum.append(1 + (boardLength+1)*i)
    return diaNegChar, diaNegNum
```

```python
def winnerCheck(board,turn,boardLength,playerMove,rowChar,colChar,diaPosChar,diaPosNum,diaNegChar,diaNegNum):
    if rowChar.count(turn) == boardLength:
        if turn == 'X':
            print("You win")
        if turn == 'O':
            print("Game Over")
        return False
    if colChar.count(turn) == boardLength:
        if turn == 'X':
            print("You win")
        if turn == 'O':
            print("Game Over")
        return False
    if diaPosChar.count('X') == boardLength:
        print("You win")
        return False
    if diaPosChar.count('O') == boardLength:
        print("Game Over")
        return False
    if diaNegChar.count('X') == boardLength:
        print("You win")
        return False
    if diaNegChar.count('O') == boardLength:
        print("Game Over")
        return False
    if list(board.values()).count('#') == 0:
        print("It's a tie")
        return False
    return True
```

```python
import random
def botMove(boardLength,playerMove,availableMoves,rowChar,rowNum,colChar,colNum,diaPosChar,diaPosNum,diaNegChar,diaNegNum):
    
    #Taking the winning move
    
    for i in range(boardLength):
        tempRowChar, tempRowNum = findRow(board,boardLength,1+(i*boardLength))
        tempColChar, tempColNum = findCol(board,boardLength,i+1)
        if tempRowChar.count('O') == boardLength - 1 and tempRowChar.count('#') == 1:
            move = tempRowNum[tempRowChar.index('#')]
            availableMoves.remove(move)
            return move, availableMoves
        if tempColChar.count('O') == boardLength - 1 and tempColChar.count('#') == 1:
            move = tempColNum[tempColChar.index('#')]
            availableMoves.remove(move)
            return move, availableMoves
    if diaPosChar.count('O') == boardLength - 1 and diaPosChar.count('#') == 1:
        move = diaPosNum[diaPosChar.index('#')]
        availableMoves.remove(move)
        return move, availableMoves
    if diaNegChar.count('O') == boardLength - 1 and diaNegChar.count('#') == 1:
        move = diaNegNum[diaNegChar.index('#')]
        availableMoves.remove(move)
        return move, availableMoves
    
    #Stopping player from winning
    
    if (rowChar.count('X') == boardLength - 1 and rowChar.count('#') == 1):# or (rowChar.count('O') == boardLength - 1 and rowChar.count('#') == 1):
        move = rowNum[rowChar.index('#')]
        availableMoves.remove(move)
        return move,availableMoves
    if (colChar.count('X') == boardLength - 1 and colChar.count('#') == 1):# or (colChar.count('O') == boardLength - 1 and colChar.count('#') == 1):
        move = colNum[colChar.index('#')]
        availableMoves.remove(move)
        return move,availableMoves
    if playerMove in diaPosNum:
        if (diaPosChar.count('X') == boardLength - 1 and diaPosChar.count('#') == 1):# or (diaPosChar.count('O') == boardLength - 1 and diaPosChar.count('#') == 1):
            move = diaPosNum[diaPosChar.index('#')]
            availableMoves.remove(move)
            return move,availableMoves
    if playerMove in diaNegNum:
        if (diaNegChar.count('X') == boardLength - 1 and diaNegChar.count('#') == 1):# or (diaNegChar.count('O') == boardLength - 1 and diaNegChar.count('#') == 1):
            move = diaNegNum[diaNegChar.index('#')]
            availableMoves.remove(move)
            return move,availableMoves
    
    # Random Move
    
    move = random.choice(availableMoves)
    availableMoves.remove(move)
    return move, availableMoves
```
